## การใช้ gTTS

```python
tts = gTTS('hello', lang='en')
tts.save('speech.mp3')
```

```python
tts = gTTS('ถึงม้วยดินสิ้นฟ้ามหาสมุทร ไม่สิ้นสุดความรักสมัครสมาน', lang='th')
tts.save('speech.mp3')
```
